```
import pandas as pd

df = pd.read_csv("./Gopfile.csv")
```

```
df.dropna(axis=0, inplace=True)

df["total"] = df['Quantity Ordered'] * df['Price Each']
df = df.groupby("Purchase Address")["total"].sum().reset_index()
```

```
import numpy as np
X = np.array(df['total']).reshape(-1, 1)
```

```
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3)
kmeans.fit(X)
```

```
df['lable'] = kmeans.labels_
```

```
def check_type (row):
    if row['lable'] == 2:
      return 'Khách VIP'
    elif row['lable'] == 1:
      return 'Khách thường'
    else:
      return 'Khách ít tương tác'
```

```
df[df['lable'] == 0]
```

```
df['Type'] = df.apply(check_type, axis=1)
df.sort_values("total", ascending=False)
```
